```python
import  numpy as np
import matplotlib.pyplot as plt                                  # gotta import the essentials so we can use numpy and matplotlib.pyplot terms

def gradient_descent(f, learning_rate, initial_point):           # we're defining the gradient descent function that python will run on our inputs
    def derivative(f, base_point):                               # estimating the derivative of the function f at the basepoint
        return (f(base_point+10**(-5))-f(base_point))/(10**(-5))

    x_coords=[initial_point]                                     # list of the x values will be stored here
    y_coords=[f(initial_point)]                                  # list of the y values will be stored here
                       
    def woo(x):                                     # we're going to use woo for the gradient descent, it's important for the absolute value function
        return (f(x+0.00001)-f(x))/0.00001
    i=1                                             #creating the variable i and assigning it a value of 1


    while abs(woo(x_coords[i-1]))>0.00000001:                     # these are the conditions we are setting for the loop to run in, the derivative 
        xi = x_coords[i-1]-learning_rate * woo(x_coords [i-1])    # has to be bigger than a certain very small number
        x_coords.append(xi)                                       # storing x and y coordinates in the list
        y_coords.append(f(x_coords[i]))

        i = i+1                                                         #to keep the loop going we increment i as x values are found closer to the
    plot_range=np.linspace(min(x_coords)-0.5, max(x_coords)+0.5, 10000) # min until their derivatives are very small and close to the min
    function_range=[f(i) for i in plot_range]
    plt.plot(plot_range, function_range)
    plt.plot(x_coords, y_coords)                                   #plotting the points (x, f(x))!

    return round(x_coords[-1],3), round(y_coords[-1],3)            #return the final x and y coordinates that are closest to the min value

def function_1(x):
    return x**2                                                 # this is our test function #1
gradient_descent(function_1, 0.9, 10)                           # we're using a learning rate of 0.9 and base point of 10 for this one
```

```python
def function_2(x):                          
    return (-3)*x**3 + 10*x**2 - 5*x - 3        # test function number 2!
gradient_descent(function_2, 0.12, 0.001)       # learning rate of 0.12 and base point of 0.001!
```

```python
def function_3(x):
    if x>0:
        return x**x
    elif x==0:
        return 1
    else:
        return abs(x)**abs(x)
        
gradient_descent(function_3, 0.88, 1.00) # it looks like there's another min there, so we'll adjust the base point value to try and find it
```

```python
gradient_descent(function_3, 0.88, -1.00)
```
