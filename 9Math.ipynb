```python
#a)first partials of (e^x)sin(y) + y^3
import sympy as sp

x, y = sp.symbols('x y')

f = (sp.exp(x))*(sp.sin(y)) + y**3 #this is the function

df_dx = sp.diff(f, x) # first partial of the function with respect to x
    
df_dy = sp.diff(f, y) # with respect to y
   
print("The first partial derivative of f with respect to x is", df_dx)
print("The first partial derivative of f with respect to y is", df_dy)
```

```python
#b) gradient vector and magnitude
import sympy as sp

x, y = sp.symbols('x y')

f = (x**2)*y + (y**2)*x

gradient = sp.Matrix([sp.diff(f, number) for number in (x, y)]) #this is finding us the matrix for the gradient

print("The gradient vector is", gradient)

point = [(x, 1), (y, -1)] #list of tuples


gradient_at_point = gradient.subs(point) # substituting values into the gradient vector

#computing the magnitude of the gradient vector at the point
magnitude = sp.sqrt(sum(part_of_vector**2 for part_of_vector in gradient_at_point))

print("Magnitude of the gradient vector at the point is:", magnitude)
```

```python
#c) Second derivates of (x^2)y + x(y^2)

import sympy as sp

x, y = sp.symbols('x y')

f = (x**2)*y + x*(y**2)

df_dx = sp.diff(f, x)
    
df_dy = sp.diff(f, y) #above this point is the same as the first cell

df_dxx = sp.diff(df_dx, x) #doing the same as the first partial, but differentiating using the
df_dxy = sp.diff(df_dx, y) # first partial (instead of original function) to get the second
    
df_dyy = sp.diff(df_dy, y) #first partial with respect to y and second with respect to y
df_dyx = sp.diff(df_dy, x) #first with respect to y and second with respect to x

print("The partial derivative of df_dx with respect to x is", df_dxx)
print("The partial derivative of df_dx with respect to y is", df_dxy)
print("The partial derivative of df_dy with respect to y is", df_dyy)
print("The partial derivative of df_dy with respect to x is", df_dyx)
```

```python
#d) Code that Greg gave us in full
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt
x, y = sp.symbols('x y')
j = x**3 - 3*x*y + y**3
j_func = sp.lambdify((x, y), j, 'numpy')
x_vals = np.linspace(-3, 3, 400)
y_vals = np.linspace(-3, 3, 400)
X, Y = np.meshgrid(x_vals, y_vals)
Z = j_func(X, Y)
plt.contourf(X, Y, Z, levels=50, cmap='viridis')
plt.colorbar()
plt.title('Contour plot of $j(x, y) = x^3 - 3xy + y^3$')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.show()
```
