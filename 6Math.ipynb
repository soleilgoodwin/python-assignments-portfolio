```python
import numpy as np
import matplotlib.pyplot as plt
def normal_density(mean, variance, x):
    coeff = 1/np.sqrt(variance*2*np.pi)
    exp_term = np.exp(-(x-mean)**2 / (2*variance))
    return coeff*exp_term
def integrating(mean, variance, a, b, num_steps=1000):
    d_x = (b-a)/num_steps
    x_values =np.linspace(a, b, num_steps)
    area = np.sum(normal_density(mean, variance, x_values)*d_x)
    return area

def distribution_plot(mean, variance, a, b, num_points=1000):
    x_coords = np.linspace(a, b, num_points)
    y_coords = normal_density(mean, variance, x_coords)

    plt.plot(x_coords, y_coords)
    plt.show()

result = integrating(171, (7.1**2), 162, 190)
distribution_plot(171, (7.1**2), 162, 190)

print("The Integrated area is:", result)
```
