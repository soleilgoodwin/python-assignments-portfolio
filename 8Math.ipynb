```python
import numpy as np
def gradient_descent(x0, y0, f, grad_f, alpha, num_iterations):
    
    x, y = x0, y0 # Initialize x and y with the initial point
    for i in range(num_iterations):
# obtain the gradient of f at (x, y):
        grad_x, grad_y = grad_f(x,y) 
# Updating x and y by taking a step in the opposite direction of the gradient (this equation for this is in the kritik):
        x = x-(alpha*grad_x)
        y = y-(alpha*grad_y)
    return x, y
```

```python
def fun_1(x,y):
    return x**2+y**2 #here's the function we'll be using
def grad_f_1(x,y):
    grad_x = 2*x #this is the partial derivative of x in the above equation
    grad_y = 2*y #this is the partial derivative of y in the above equation
    return grad_x, grad_y #these values are the gradients at x and y 



print (gradient_descent(0.1,0.1,fun_1, grad_f_1, 0.1, 10)) #testing the code with these inputs from the kritik
print (gradient_descent(-1,1,fun_1, grad_f_1, 0.01, 100))
```

```python
import numpy as np
def fun_2(x,y):
    return 1-np.exp(-x**2-(y-2)**2)-2*np.exp(-x**2-(y+2)**2) #new equation
def grad_f_2(x,y):
    h = 0.00001
    grad_x = (fun_2(x+h,y)-fun_2(x,y))/h  # using the formal definition of a limit here so we don't have to calculate the derivatives manually
    grad_y = (fun_2(x, y+h)-fun_2(x,y))/h
    return grad_x, grad_y #value of the gradient at x and y again but for our new equation
print(gradient_descent(0,1,fun_2, grad_f_2, 0.01, 10000))

import numpy as np
from mpl_toolkits import mplot3d #for 3D plots
import matplotlib.pyplot as plt #the normal 

X=np.linspace(-5,5,100)
Y=np.linspace(-5,5,100)
x,y=np.meshgrid(X,Y)
z=fun_2(x,y)

fig = plt.figure()
ax = plt.axes (projection='3d')
ax.plot_surface(x,y,z, cmap='viridis', edgecolor='none')
```
